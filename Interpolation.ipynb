# Interpolation

Firstly, using the Gaussian distribution

$$
P(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi}}\frac{e^{-\frac{\left(x-\mu\right)^2}{2\mu^2}}}{\sigma},
$$

and

$$
C(x) \propto Erf.
$$

Calculate the interpolated version of $P(x|\mu, \sigma)$ and $C(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import stats_funcs.funcs as fs
from random import uniform
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
```

```python
mu = 0
sigma = 1

x = np.linspace(-4, 4, 50)

pdf = [fs.gaussian(i, mu, sigma) for i in x]

x_inter = fs.interpolate(x, x)
pdf_inter = fs.interpolate(pdf, x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Interpolation Comparisson')
ax1.bar(x, pdf, alpha = 0.4, label = 'Original $\{x_i\}$ with ' + f'N={len(pdf)} values.')
ax1.legend()
ax2.bar(x_inter, pdf_inter, alpha = 0.4, label = 'Interpolated $\{x_i\}$ with ' + f'N={len(pdf_inter)} values.')
ax2.legend()
```

```python
erf = fs.cdf(fs.gaussian, x, (mu, sigma))
erf_inter = fs.interpolate(erf, x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Interpolation Comparisson')
ax1.bar(x, erf, alpha = 0.4, label = 'Original $\{x_i\}$ with ' + f'N={len(erf)} values.')
ax1.legend()
ax2.bar(x_inter, erf_inter, alpha = 0.4, label = 'Interpolated $\{x_i\}$ with ' + f'N={len(erf_inter)} values.')
ax2.legend()
```
